## The data set I will be loading in concerns covid-19 vaccinations rates among different demographics in the US
### URL: https://data.cdc.gov/Vaccinations/COVID-19-Vaccination-Demographics-in-the-United-St/km4m-vcsb/data

#### What I'm curious about is the percentage of boosters shots administered among various demographics. I'm also interested in the number of 1st doses administered among different demographics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#pd.set_option('display.max_rows', None) # or 1000.
#pd.reset_option("^display")


df = pd.read_csv('COVID-19_Vaccination_Demographics_in_the_United_States_National.csv')

df.describe()
#display(df)
```

```python
print(df.columns)
print(df.columns.array)
```

```python
table = pd.pivot_table(df, values='Booster_Doses_Pop_Pct_known', index=['Demographic_category'],
                    columns=[], aggfunc = np.mean)
table.plot.bar(figsize = (15,10))
plt.title('Percentage of Booster Doses Administered by Demographics')
plt.ylabel('Percentage (%)')

#mean of 1 dose administered
table2 = pd.pivot_table(df, values='Administered_Dose1_pct_US', index=['Demographic_category'],
                    columns=[], aggfunc = np.mean)
table2.plot.bar(figsize = (15,10))
plt.title('Percentage of Administered First Dose in US by Various Demographics')
plt.ylabel('Percentage (%)')

#group by demographic category and sum
grouped_df = df.groupby(['Demographic_category']).mean()
grouped_df = grouped_df[['Series_Complete_Yes' , 'Booster_Doses_Yes']]

#display(grouped_df)
grouped_df.plot.bar(figsize = (15,10))
plt.title('Count of US people who finished the series and got the booster among different demographics')
plt.ylabel('Count')
plt.show()
```

In my exploration, we can see that generally the age group 50-75+ have the higher rates for getting the first dose and getting the booster shot. I imagined that if people in this age group had gotten the first dose, then they're more likely to receieve the booster and the data does seem to show that correlation slightly. In the percentage of booster shots administered bar chart, ages 50-75+ has the highest rates for getting the booster.
In the third bar chart, i was simply curious to see how many people have finished the series among different demographics(which i believe is getting 2 shots). In total, 140 million people finished the 2 dose series and 20 million people have received the booster
